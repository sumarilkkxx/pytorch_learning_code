```python
# 导入常用的类
from torch.utils.tensorboard import SummaryWriter
from PIL import Image
import numpy as np
```

```python
# 创建一个实例
writer =  SummaryWriter("logs")
# writer.add_image()
# 使用简单函数测试绘图功能 y = x
for i in range(100):
    writer.add_scalar("y = x" ,i ,i)
writer.close()
```

```python
# 查看文件的格式是否满足要求，在add_image方法中导入的图片需为tensor形式或者numpy形式
image_path = "../dataset/train/ants_image/0013035.jpg"
img_test = Image.open(image_path)
# img_test.show()\
print(type(img_test))
```

```python
img_array = np.array(img_test)
img_array.shape
```

```python
#注意原有的图像矩阵为(H,W,C)格式，但是add_image方法默认支持CHW格式，因此，在此需要指定数据格式为“HWC”，否则会有报错
writer.add_image("test",img_array,global_step=1, dataformats="HWC")
```

```python
#使用tensorboard查看写入的图片，命令为 tensorboard --logdir =logs
#global_step的含义为输入给模型的每一步图像，与前面的tag标签绑定 
```
